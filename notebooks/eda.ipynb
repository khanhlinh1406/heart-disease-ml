```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/heart.csv")

print("head: \n")
print(df.head())

print("info: \n")
df.info()

print("describe: \n")
df.describe()

print("Check missing value: \n")
print(df.isnull().sum())

print("Check distribution: \n")
df["age"].hist()
#df["sex"].hist()

print("CORRELATION: \n")
sns.heatmap(df.corr(), annot=True)

print("RELATIONSHIP: \n")
sns.scatterplot(x="age", y="chol", hue="target", data=df)
```
